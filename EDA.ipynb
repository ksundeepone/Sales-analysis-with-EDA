```python


# Assuming 'combined_df' is your DataFrame and 'Quantity Ordered' is the column you want to check
non_numeric_mask = pd.to_numeric(combined_df['Quantity Ordered'], errors='coerce').isna()
non_numeric_values = combined_df.loc[non_numeric_mask, 'Quantity Ordered']

# Display unique non-numeric values
print(non_numeric_values.unique())
```

```python
combined_df['Quantity Ordered'] = pd.to_numeric(combined_df['Quantity Ordered'], errors='coerce')
```

```python
combined_df['Quantity Ordered'] = combined_df['Quantity Ordered'].fillna(0).astype('int32')



```

```python
#combined_df['Order ID'] = pd.to_numeric(combined_df['Order ID'].fillna(0).astype('int32'))

combined_df['Order ID'] = pd.to_numeric(combined_df['Order ID'], errors='coerce').astype('Int32')
```

```python
combined_df['Order Date'] = pd.to_datetime(combined_df['Order Date'], errors = 'coerce')
```

```python
combined_df.dtypes
```

```python
combined_df['Price Each']= pd.to_numeric(combined_df['Price Each'], errors = 'coerce').astype('float')
```

```python
combined_df['month'] = combined_df['Order Date'].dt.month
```

```python
combined_df['month'] = (combined_df['Order Date'].dt.month).astype('Int32')
```

```python
combined_df.head()
```

```python
combined_df['day'] = (combined_df['Order Date'].dt.day).astype('Int32')
combined_df['year'] = (combined_df['Order Date'].dt.year).astype('Int32')
combined_df.head()
```

```python
combined_df['sales'] = combined_df['Price Each'] * combined_df['Quantity Ordered']

combined_df.head()
```

```python
from yellowbrick.features import Rank2D
import pandas as pd

# Load your DataFrame
df = pd.read_csv('/Users/sundeep/Downloads/UTA/DATASETS/Life_Expectancy_Data.csv')

# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Extract the numeric features from the DataFrame
X = df[numeric_columns]

# Instantiate the Rank2D visualizer
visualizer = Rank2D(algorithm='pearson')  # You can change the algorithm as needed

# Fit and transform the data
visualizer.fit_transform(X)

# Display the visualization
visualizer.show()
```

```python
from yellowbrick.features import parallel_coordinates
import pandas as pd

# Load your DataFrame
df = pd.read_csv('/Users/sundeep/Downloads/UTA/DATASETS/Life_Expectancy_Data.csv')

# Specify the target column (the one you want to predict)
target_column = 'Alcohol'

# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Exclude the target column from features
numeric_columns = [col for col in numeric_columns if col != target_column]

# Extract features (X) and target (y)
X = df[numeric_columns]
y = df['Alcohol']

# Create and display the parallel coordinates visualization
visualizer = parallel_coordinates(X, y, normalize="standard")
visualizer.show()
```

```python
from yellowbrick.features import RadViz
import pandas as pd

# Load your DataFrame
df = pd.read_csv('/Users/sundeep/Downloads/UTA/DATASETS/Life_Expectancy_Data.csv')

# Assuming 'target_column' is the column you want to predict
target_column = 'Alcohol'

# Extract features (X) and target (y)
X = df.drop(target_column, axis=1)  # Exclude the target column from features
y = df['Alcohol']

# Create and display the RadViz visualization
visualizer = RadViz(colors=["maroon", "gold"])
visualizer.fit_transform(X, y)
visualizer.show()
```
